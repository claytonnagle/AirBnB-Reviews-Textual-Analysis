```python
copy_lisa_df = selected_lisa.copy()
```

```python
# pre-proessing text 
# object is a string
copy_lisa_df['comments'] = copy_lisa_df['comments'].astype(str)
copy_lisa_df.dtypes
```

```python
copy_lisa_df['comments'][0]
```

```python
copy_lisa_df['comments'] = copy_lisa_df['comments'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
# lowercase 
copy_lisa_df['comments'][0]
```

```python
# remove punctuation 
copy_lisa_df['comments'] = copy_lisa_df['comments'].str.replace(r'[^\w\s]','')
copy_lisa_df['comments'][0]
```

```python
# stop words 
stop = stopwords.words('english')
copy_lisa_df['comments'] = copy_lisa_df['comments'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
copy_lisa_df['comments'][0]
```

```python
readable_lisa = copy_lisa_df['comments']
readable_lisa[0]
```

```python
# delete suffixes
suffix = PorterStemmer()
copy_lisa_df['comments'] = copy_lisa_df['comments'].apply(lambda x: " ".join([suffix.stem(word) for word in x.split()]))
copy_lisa_df['comments'][0]
```

```python
def sentimentFunc(x):
    return TextBlob(x).sentiment
```

```python
copy_lisa_df['score'] = copy_lisa_df['comments'].apply(sentimentFunc)
copy_lisa_df.head(20)

# polarity (1st score) assuming >.0 is positive [-1.0, 1.0]
# subjectity (2nd score) assuming >.50 is opinion (subjective) than factual information
```

```python
sentiment_category = []
```

```python
for score in range(len(polarity_scores)):
    if polarity_scores[score] >= .50:
        sentiment_category.append('morepositve')
    elif polarity_scores[score] <.50 and polarity_scores[score] >0:
        sentiment_category.append('lesspositive')
    elif polarity_scores[score] <0 and polarity_scores[score] >-.50:
        sentiment_category.append('lessnegative')
    elif polarity_scores[score] <-.50:
        sentiment_category.append('morenegative')
    else:
        sentiment_category.append('neutral')
```

```python
len(sentiment_category)
```

```python
clean_df = copy_lisa_df.copy()
clean_df
```

```python
clean_df["polarity score"] = polarity_scores
clean_df["sentiment category"] = sentiment_category
```

```python
clean_df
```

```python
classify_df =  pd.get_dummies(clean_df)
classify_df.columns
```

```python
clean_df['lesspostive']= classify_df['sentiment category_lesspositive']
clean_df['morepostive']= classify_df['sentiment category_morepositve']
clean_df['neutral']= classify_df['sentiment category_neutral']
```

```python
# look for positive comments 
morepostive_df = clean_df.loc[clean_df['sentiment category'] == 'morepositve']
morepostive_df
```
